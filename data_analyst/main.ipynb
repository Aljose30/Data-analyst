# ETL

Antes de procesar datos, se limpian y normalizan los datos para que sea mas facil el analisis

```python
#eliminando columnas que se consideran no relevantes para el analisis

coursera_courses.drop(['course_url', 'institution'], axis=1, inplace=True)
coursera_reviews.drop(['reviewers', 'reviews'], axis=1, inplace=True)
udemy_courses.drop(['course_id','url','is_paid'], axis=1, inplace=True)
edx_courses.drop(['summary','institution','instructors','subtitles','course_effort', 'course_description','course_url'], axis=1, inplace=True)
```

```python
edx_courses.head(5)
```

```python
#se hacen transformaciones a ciertas columnas del dataset Edx
# Separación de la columna 'price' en dos columnas
edx_courses['extra'], edx_courses['Price[USD]'] = edx_courses['price'].str.split('$', 1).str
edx_courses['Price[USD]'], edx_courses['extra1'] = edx_courses['Price[USD]'].str.split(' ', 1).str

# Eliminación de columnas no necesarias
edx_courses.drop(['extra', 'extra1'], axis=1, inplace=True)

# Separación de la columna 'course_length' en dos columnas
edx_courses['Length[Weeks]'], edx_courses['extra'] = edx_courses['course_length'].str.split(' ', 1).str

# Eliminación de columnas no necesarias
edx_courses.drop(['extra', 'course_length'], axis=1, inplace=True)

# Modificación de los títulos de las columnas
cols_rename = {
    'title': 'Title',
    'n_enrolled': 'Enrolled',
    'course_type': 'Course_Type',
    'subject': 'Subject',
    'language': 'Language',
    'price': 'Price[USD]'
}
edx_courses.rename(cols_rename, axis=1, inplace=True)
```

```python
edx_courses.head(5)
```

```python
#Unir los datasets coursera courses/renews y se modifican los titulos 

# Agrupación por promedio en la tabla coursera_r
group_cols = ['course_id']
coursera_reviews = coursera_reviews.groupby(group_cols).mean()

# Unificación de las tablas coursera_courses y coursera_renew en una sola tabla 'coursera'
merge_cols = ['course_id']
coursera = coursera_courses.merge(coursera_reviews, on=merge_cols, how='left')

# Eliminación de la columna 'course_id'
drop_cols = ['course_id']
coursera.drop(drop_cols, axis=1, inplace=True)

# Modificación de los títulos de las columnas
cols_rename = {
    'name': 'Title',
    'rating': 'Rating'
}
coursera.rename(cols_rename, axis=1, inplace=True)
```
